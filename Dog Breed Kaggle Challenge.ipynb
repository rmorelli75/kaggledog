```python
def rgb (colors):
    return (colors[0]+colors[1]+colors[2])/3.0

def rgb_to_greyscale(input_image):
    output_image=[]
    for j in input_image:
        temp_xlist=[]
        for k in j:
            temp_xlist.append(rgb(k))
        output_image.append(temp_xlist)
    return output_image
            
```
